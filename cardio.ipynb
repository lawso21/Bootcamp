# Cardio Good Fitness

## Identifying the market for 3 different treadmills

### Hypothesis: Different groups of purchaser might prefer different products

#### Read in the data from 'CardioGoodFitness.csv' and have a first look

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('CardioGoodFitness.csv')
```

## Data looks clean. No missing values

### Quick look at the 3 products versus the other numerical data to see if there is any noticeable difference

```python
columns = ['Age','Education','Usage','Fitness','Income','Miles']

for column in columns:
    df.plot(x='Product', y=column,kind='scatter',title=(f' Product by {column}'));
```

## Products TM195 and TM498 appeal to similar customers
## TM798 appeals to older, wealthier and more committed runners

```python
df_grouped = df.groupby(by=['Product','Gender']).size().unstack()

df_grouped
```

### TM798 does not appeal to Females
### Could be missing opportunities with Females

```python
df_grouped.plot(kind='bar',stacked=True, color=['tomato','royalblue'])
plt.title('Product sales by gender')
plt.xticks(rotation=0, ha='center')
plt.ylabel('Number of sales')
```

## TM195 is the most popular treadmill 

```python
product_sales = df.groupby(by=['Product']).size()

product_sales
```

```python
product_sales.plot(kind='pie',title='Product sales', colors=['royalblue','gold','limegreen'])
```

```python
df_income = df.groupby(by=['Product'])['Income'].mean().round(1)

df_income
```

```python
df_income.plot(kind='bar',color=['green','blue','darkturquoise'], title='Average Income of purchasers')
plt.ylabel('Average Income') 
plt.xticks(rotation=0)
```

## TM798 is bought by those with higher incomes
### Recommend efforts to increase sales of TM798 to Females as possible untapped market.
